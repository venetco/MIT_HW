```python
# ============================================================
#  CREDIT SCORING ML PIPELINE
#  UCI Credit Approval Dataset (ID=27)
#  Features: reverse-engineered names from anonymized data
# ============================================================

# ── CELL 1: Install dependencies ─────────────────────────────
# !pip install ucimlrepo xgboost shap --quiet

# ── CELL 2: Imports ───────────────────────────────────────────
import warnings; warnings.filterwarnings("ignore")
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import shap
from sklearn.model_selection import StratifiedKFold, cross_validate, train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OrdinalEncoder
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import (
    roc_auc_score, roc_curve, precision_recall_curve,
    classification_report, confusion_matrix,
    average_precision_score, ConfusionMatrixDisplay
)
from xgboost import XGBClassifier
from ucimlrepo import fetch_ucirepo

# ── White plot style ──────────────────────────────────────────
plt.rcParams.update({
    "figure.facecolor":  "white",
    "axes.facecolor":    "white",
    "axes.edgecolor":    "#CCCCCC",
    "axes.spines.top":   False,
    "axes.spines.right": False,
    "text.color":        "#1A1A2E",
    "axes.labelcolor":   "#1A1A2E",
    "xtick.color":       "#555555",
    "ytick.color":       "#555555",
    "grid.color":        "#EEEEEE",
    "grid.linestyle":    "--",
    "grid.alpha":        0.7,
    "font.family":       "DejaVu Sans",
    "axes.titlesize":    12,
    "axes.labelsize":    10,
})

NAVY   = "#1A1A2E"
BLUE   = "#1F77B4"
GREEN  = "#2CA02C"
ORANGE = "#FF7F0E"
RED    = "#D62728"
PURPLE = "#9467BD"
TEAL   = "#17BECF"
COLORS = [BLUE, GREEN, ORANGE, RED, PURPLE]

# ── CELL 3: Load Dataset ──────────────────────────────────────
print("Loading UCI Credit Approval Dataset (ID=27)...")
credit_approval = fetch_ucirepo(id=27)
X_raw = credit_approval.data.features.copy()
y_raw = credit_approval.data.targets.copy()

target_col = y_raw.columns[0]
y = (y_raw[target_col] == "+").astype(int)

print(f"  Rows         : {len(X_raw)}")
print(f"  Features     : {X_raw.shape[1]}")
print(f"  Approval rate: {y.mean():.1%}")
print(f"  Missing vals : {X_raw.isnull().sum().sum()}")

# ── CELL 4: Rename features to meaningful names ───────────────
# The UCI dataset deliberately anonymized all column names.
# The mapping below is the consensus reverse-engineering from
# the research community based on data types and domain knowledge.
FEATURE_RENAME = {
    "A1":  "Gender",
    "A2":  "Age",
    "A3":  "Debt",
    "A4":  "Marital_Status",
    "A5":  "Bank_Customer_Type",
    "A6":  "Education_Level",
    "A7":  "Ethnicity",
    "A8":  "Years_Employed",
    "A9":  "Prior_Default",
    "A10": "Employed",
    "A11": "Credit_Score",
    "A12": "Drivers_License",
    "A13": "Citizenship",
    "A14": "Zip_Code",
    "A15": "Income",
}

# Value label mappings for categorical features
VALUE_LABELS = {
    "Gender":           {"b": "Male",  "a": "Female"},
    "Marital_Status":   {"u": "Married", "y": "Single", "l": "Divorced", "t": "Separated"},
    "Bank_Customer_Type": {"g": "Standard", "p": "Premium", "gg": "Basic"},
    "Prior_Default":    {"t": "Yes", "f": "No"},
    "Employed":         {"t": "Yes", "f": "No"},
    "Drivers_License":  {"t": "Yes", "f": "No"},
    "Citizenship":      {"g": "Citizen", "p": "Permanent_Resident", "s": "Temporary"},
}

X = X_raw.rename(columns=FEATURE_RENAME)

# Apply value labels
for col, mapping in VALUE_LABELS.items():
    if col in X.columns:
        X[col] = X[col].map(mapping).fillna(X[col])   # keep unknowns as-is

print("\nRenamed feature sample:")
print(X.head(3).to_string())

# ── CELL 5: Preprocessing ─────────────────────────────────────
cat_cols = X.select_dtypes(include="object").columns.tolist()
num_cols = X.select_dtypes(exclude="object").columns.tolist()

print(f"\nNumerical features  ({len(num_cols)}): {num_cols}")
print(f"Categorical features({len(cat_cols)}): {cat_cols}")

num_pipe = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler",  StandardScaler()),
])
cat_pipe = Pipeline([
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("encoder", OrdinalEncoder(handle_unknown="use_encoded_value", unknown_value=-1)),
])
preprocessor = ColumnTransformer([
    ("num", num_pipe, num_cols),
    ("cat", cat_pipe, cat_cols),
])

X_proc = preprocessor.fit_transform(X)
feature_names = num_cols + cat_cols   # used for SHAP labels

X_train, X_test, y_train, y_test = train_test_split(
    X_proc, y, test_size=0.2, random_state=42, stratify=y
)
print(f"\nTrain size: {len(X_train)}  |  Test size: {len(X_test)}")

# ── CELL 6: EDA ───────────────────────────────────────────────
fig, axes = plt.subplots(3, 5, figsize=(18, 11))
fig.suptitle("Exploratory Data Analysis — UCI Credit Approval",
             fontsize=15, fontweight="bold", color=NAVY, y=1.01)

all_cols = num_cols + cat_cols
for ax, col in zip(axes.flat, all_cols):
    if col in num_cols:
        approved = X[col][y == 1].dropna()
        denied   = X[col][y == 0].dropna()
        ax.hist(denied,   bins=25, alpha=0.6, color=RED,   label="Denied",   density=True)
        ax.hist(approved, bins=25, alpha=0.6, color=GREEN, label="Approved", density=True)
        ax.set_title(col, fontweight="bold", fontsize=9)
        ax.legend(fontsize=6)
    else:
        vc = X.groupby([col, y.rename("Target")]).size().unstack(fill_value=0)
        vc.columns = ["Denied", "Approved"]
        vc_pct = vc.div(vc.sum(axis=1), axis=0) * 100
        vc_pct.plot(kind="bar", ax=ax, color=[RED, GREEN], alpha=0.8, width=0.7)
        ax.set_title(col, fontweight="bold", fontsize=9)
        ax.set_xticklabels(ax.get_xticklabels(), rotation=30, fontsize=7)
        ax.legend(fontsize=6)
    ax.grid(axis="y", alpha=0.5)

# Hide unused subplots
for ax in axes.flat[len(all_cols):]:
    ax.set_visible(False)

plt.tight_layout()
plt.savefig("fig0_eda.png", dpi=150, bbox_inches="tight", facecolor="white")
plt.show()
print("[Saved] fig0_eda.png")

# ── CELL 7: Model Training & 5-Fold Cross-Validation ─────────
models = {
    "Logistic\nRegression":          LogisticRegression(max_iter=1000, random_state=42),
    "Random\nForest":                RandomForestClassifier(n_estimators=300, random_state=42),
    "XGBoost\nClassification":       XGBClassifier(n_estimators=300, eval_metric="logloss",
                                                    random_state=42, verbosity=0),
    "Support Vector\nClassification":SVC(probability=True, kernel="rbf", random_state=42),
    "Neural Network\nClassification":MLPClassifier(hidden_layer_sizes=(64, 32),
                                                    max_iter=500, random_state=42),
}

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_results = {}

print("\n5-Fold Cross-Validation Results:")
print(f"{'Model':35s}  {'Accuracy':>12s}  {'AUC-ROC':>12s}  {'F1':>10s}")
print("-" * 75)

for name, model in models.items():
    scores = cross_validate(
        model, X_proc, y, cv=cv,
        scoring=["accuracy", "roc_auc", "average_precision", "f1"],
    )
    cv_results[name] = {k: scores[f"test_{k}"] for k in ["accuracy","roc_auc","average_precision","f1"]}
    flat = name.replace("\n", " ")
    print(f"{flat:35s}  "
          f"{scores['test_accuracy'].mean():.3f} ± {scores['test_accuracy'].std():.3f}  "
          f"{scores['test_roc_auc'].mean():.3f} ± {scores['test_roc_auc'].std():.3f}  "
          f"{scores['test_f1'].mean():.3f} ± {scores['test_f1'].std():.3f}")

# ── CELL 8: Fig 1 — Model Comparison Chart ───────────────────
fig, axes = plt.subplots(1, 2, figsize=(14, 6))
fig.suptitle("Baseline Model Performance — 5-Fold Cross-Validation",
             fontsize=14, fontweight="bold", color=NAVY)

model_names = list(cv_results.keys())

for ax, (mk, ml) in zip(axes, [("accuracy", "Accuracy (%)"), ("roc_auc", "ROC-AUC (%)")]):
    means = [cv_results[m][mk].mean() * 100 for m in model_names]
    stds  = [cv_results[m][mk].std()  * 100 for m in model_names]

    for i, (mean, std) in enumerate(zip(means, stds)):
        ax.errorbar(mean, i, xerr=std, fmt="o", color=COLORS[i],
                    capsize=6, capthick=2, elinewidth=2.5, markersize=11,
                    markeredgecolor="white", markeredgewidth=1.5)
        ax.text(mean + std + 0.3, i, f"{mean:.1f}%", va="center",
                fontsize=8.5, color=COLORS[i], fontweight="bold")

    ax.set_yticks(range(len(model_names)))
    ax.set_yticklabels(model_names, fontsize=9)
    ax.set_xlabel(ml, fontsize=10)
    ax.set_title(ml.split()[0], fontsize=11, fontweight="bold", color=NAVY)
    ax.set_xlim(50, 102)
    ax.grid(axis="x", alpha=0.5)
    ax.axvline(means[np.argmax(means)], color=COLORS[np.argmax(means)],
               linestyle=":", alpha=0.4)

plt.tight_layout()
plt.savefig("fig1_model_comparison.png", dpi=150, bbox_inches="tight", facecolor="white")
plt.show()
print("[Saved] fig1_model_comparison.png")

# ── CELL 9: Champion Model — Full Evaluation ─────────────────
print("\nFitting champion model (XGBoost)...")
champion = XGBClassifier(n_estimators=300, eval_metric="logloss",
                         random_state=42, verbosity=0)
champion.fit(X_train, y_train)
y_prob = champion.predict_proba(X_test)[:, 1]
y_pred = champion.predict(X_test)

auc  = roc_auc_score(y_test, y_prob)
ap   = average_precision_score(y_test, y_prob)
fpr, tpr, thr = roc_curve(y_test, y_prob)
ks_stat = max(tpr - fpr)
ks_thr  = thr[np.argmax(tpr - fpr)]

print(f"\n{'─'*45}")
print(f"  Test AUC-ROC        : {auc:.4f}")
print(f"  Gini Coefficient    : {2*auc - 1:.4f}")
print(f"  KS Statistic        : {ks_stat:.4f}  (at threshold {ks_thr:.3f})")
print(f"  Avg Precision (PR)  : {ap:.4f}")
print(f"{'─'*45}")
print(f"\n{classification_report(y_test, y_pred, target_names=['Denied','Approved'])}")

# ── CELL 10: Fig 2 — Evaluation Plots ────────────────────────
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
fig.suptitle("XGBoost Champion — Hold-Out Test Evaluation",
             fontsize=14, fontweight="bold", color=NAVY)

# ROC Curve
axes[0].plot(fpr, tpr, color=BLUE, lw=2.5, label=f"XGBoost  AUC = {auc:.3f}")
axes[0].plot([0, 1], [0, 1], "--", color="#AAAAAA", lw=1.5, label="Random (AUC = 0.500)")
axes[0].axvline(fpr[np.argmax(tpr - fpr)], color=ORANGE, linestyle=":",
                lw=2, label=f"KS = {ks_stat:.3f}")
axes[0].fill_between(fpr, tpr, alpha=0.08, color=BLUE)
axes[0].set(xlabel="False Positive Rate", ylabel="True Positive Rate", title="ROC Curve")
axes[0].legend(fontsize=8.5)

# Precision-Recall
prec, rec, _ = precision_recall_curve(y_test, y_prob)
baseline_pr  = y_test.mean()
axes[1].plot(rec, prec, color=GREEN, lw=2.5, label=f"XGBoost  AP = {ap:.3f}")
axes[1].axhline(baseline_pr, color="#AAAAAA", linestyle="--", lw=1.5,
                label=f"Baseline = {baseline_pr:.3f}")
axes[1].fill_between(rec, prec, alpha=0.08, color=GREEN)
axes[1].set(xlabel="Recall", ylabel="Precision", title="Precision-Recall Curve")
axes[1].legend(fontsize=8.5)

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(cm, display_labels=["Denied", "Approved"])
disp.plot(ax=axes[2], colorbar=False, cmap="Blues")
axes[2].set_title("Confusion Matrix")

plt.tight_layout()
plt.savefig("fig2_evaluation.png", dpi=150, bbox_inches="tight", facecolor="white")
plt.show()
print("[Saved] fig2_evaluation.png")

# ── CELL 11: SHAP Explainability ──────────────────────────────
print("\nComputing SHAP values...")
explainer   = shap.TreeExplainer(champion)
shap_values = explainer(X_test)

# Rename SHAP feature axis for readability
shap_values.feature_names = feature_names

# Fig 3a: Beeswarm (global importance)
fig, ax = plt.subplots(figsize=(10, 6))
fig.patch.set_facecolor("white")
shap.plots.beeswarm(shap_values, max_display=15, show=False, color_bar=True)
plt.title("SHAP Feature Importance — Which Features Drive Credit Decisions?",
          fontsize=12, fontweight="bold", color=NAVY, pad=14)
plt.tight_layout()
plt.savefig("fig3a_shap_beeswarm.png", dpi=150, bbox_inches="tight", facecolor="white")
plt.show()
print("[Saved] fig3a_shap_beeswarm.png")

# Fig 3b: Bar summary
fig, ax = plt.subplots(figsize=(9, 5))
fig.patch.set_facecolor("white")
shap.plots.bar(shap_values, max_display=15, show=False)
plt.title("Mean |SHAP| — Average Feature Impact",
          fontsize=12, fontweight="bold", color=NAVY, pad=14)
plt.tight_layout()
plt.savefig("fig3b_shap_bar.png", dpi=150, bbox_inches="tight", facecolor="white")
plt.show()
print("[Saved] fig3b_shap_bar.png")

# ── CELL 12: Individual SHAP Waterfall (Adverse Action Notices) ──
def plot_waterfall_with_notice(idx):
    """Plot SHAP waterfall + print bank-style adverse action notice."""
    pred_prob = y_prob[idx]
    true_lbl  = "Approved ✓" if y_test.iloc[idx] == 1 else "Denied ✗"
    decision  = "APPROVED" if pred_prob >= 0.5 else "DECLINED"

    fig, ax = plt.subplots(figsize=(11, 6))
    fig.patch.set_facecolor("white")
    shap.plots.waterfall(shap_values[idx], max_display=12, show=False)
    plt.title(
        f"Individual Credit Decision — Applicant #{idx}\n"
        f"P(Approve) = {pred_prob:.1%}   |   Decision: {decision}   |   Actual: {true_lbl}",
        fontsize=11, fontweight="bold", color=NAVY, pad=12
    )
    plt.tight_layout()
    fname = f"fig4_waterfall_applicant_{idx}.png"
    plt.savefig(fname, dpi=150, bbox_inches="tight", facecolor="white")
    plt.show()

    # Adverse action notice
    sv_ranked  = sorted(zip(shap_values[idx].values, feature_names))
    top_adverse = sv_ranked[:3]   # most negative (driving denial)
    print(f"""
┌─────────────────────────────────────────────────────────────┐
│             ADVERSE ACTION NOTICE                           │
│    (Fair Credit Reporting Act — 15 U.S.C. § 1681m)         │
├─────────────────────────────────────────────────────────────┤
│  Applicant ID  : {str(idx).ljust(10)}                               │
│  Decision      : {decision.ljust(10)}  (Model P={pred_prob:.1%})           │
│  Actual Label  : {true_lbl.ljust(12)}                             │
├─────────────────────────────────────────────────────────────┤
│  PRIMARY ADVERSE REASON CODES                               │
│  (Factors that most negatively impacted the decision)       │""")
    reasons = ["Insufficient income or employment history",
               "Negative prior credit history",
               "High debt-to-income ratio",
               "Short credit history",
               "Insufficient collateral"]
    for i, (sv_val, fname_feat) in enumerate(top_adverse, 1):
        print(f"│  {i}. {fname_feat:<20s}  (SHAP = {sv_val:+.3f})                   │")
    print("└─────────────────────────────────────────────────────────────┘\n")

# Run for 3 applicants with different outcomes
for i in [0, 5, 15]:
    plot_waterfall_with_notice(i)

# ── CELL 13: Credit Scorecard (300-850 Scale) ─────────────────
PDO    = 20             # Points to Double Odds (industry standard)
OFFSET = 600            # Score at base odds 1:1
FACTOR = PDO / np.log(2)

p_safe = np.clip(y_prob, 1e-6, 1 - 1e-6)
log_odds = np.log(p_safe / (1 - p_safe))
scores   = np.clip(OFFSET + FACTOR * (-log_odds), 300, 850).round().astype(int)

# Score band analysis
band_labels = ["Very Poor\n300–499", "Fair\n500–579", "Good\n580–669",
               "Very Good\n670–739", "Exceptional\n740–799", "Super Prime\n800–850"]
bands = pd.cut(scores, bins=[300, 500, 580, 670, 740, 800, 850], labels=band_labels)
band_df = pd.DataFrame({"score": scores, "band": bands,
                         "actual": y_test.values, "prob": y_prob})
band_stats = band_df.groupby("band", observed=False).agg(
    count=("score", "count"),
    approval_rate=("actual", "mean"),
    avg_prob=("prob", "mean"),
).round(3)

print("\n── Credit Score Band Analysis ──────────────────────────────")
print(band_stats.to_string())
print(f"\n  Score range : {scores.min()} – {scores.max()}")
print(f"  Mean score  : {scores.mean():.0f}")
print(f"  Median score: {np.median(scores):.0f}")

# Fig 5: Scorecard plots
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle("Credit Scorecard (300–850 Scale, PDO=20)",
             fontsize=14, fontweight="bold", color=NAVY)

# Score distribution
bins = np.linspace(300, 850, 40)
axes[0].hist(scores[y_test.values == 0], bins=bins, alpha=0.65, color=RED,
             label="Denied", density=True, edgecolor="white")
axes[0].hist(scores[y_test.values == 1], bins=bins, alpha=0.65, color=GREEN,
             label="Approved", density=True, edgecolor="white")
axes[0].axvline(500, color="#AAAAAA", linestyle=":", lw=1.5)
axes[0].set(xlabel="Credit Score", ylabel="Density", title="Score Distribution by Outcome")
axes[0].legend()

# Approval rate by band
band_colors = [RED, ORANGE, "#F0C800", GREEN, BLUE, PURPLE]
valid_bands = band_stats[band_stats["count"] > 0]
bars = axes[1].bar(range(len(valid_bands)), valid_bands["approval_rate"] * 100,
                   color=band_colors[:len(valid_bands)], alpha=0.85, edgecolor="white")
axes[1].set_xticks(range(len(valid_bands)))
axes[1].set_xticklabels(valid_bands.index, fontsize=7.5)
axes[1].set(ylabel="Approval Rate (%)", title="Approval Rate by Score Band", ylim=(0, 115))
for bar, (_, row) in zip(bars, valid_bands.iterrows()):
    axes[1].text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 2,
                 f"n={int(row['count'])}\n{row['approval_rate']:.0%}",
                 ha="center", fontsize=8, fontweight="bold")

# Score vs Probability (calibration)
axes[2].scatter(scores, y_prob, alpha=0.5, c=y_test.values,
                cmap="RdYlGn", s=40, edgecolors="white", lw=0.5)
axes[2].axhline(0.5, color="#AAAAAA", linestyle="--", lw=1.5)
axes[2].set(xlabel="Credit Score", ylabel="P(Approve) from Model",
            title="Score vs Model Probability")
sm = plt.cm.ScalarMappable(cmap="RdYlGn", norm=plt.Normalize(0, 1))
plt.colorbar(sm, ax=axes[2], label="Actual (0=Denied, 1=Approved)")

plt.tight_layout()
plt.savefig("fig5_scorecard.png", dpi=150, bbox_inches="tight", facecolor="white")
plt.show()
print("[Saved] fig5_scorecard.png")

# ── CELL 14: Financial Backtest ───────────────────────────────
# Portfolio assumptions (adjustable)
LOAN_AMOUNT   = 10_000   # $ average loan
INTEREST_RATE = 0.12     # 12% annual interest
RECOVERY_RATE = 0.40     # 40% recovery on defaulted loans
N_PORTFOLIO   = 10_000   # hypothetical portfolio size

revenue_per_good_loan = LOAN_AMOUNT * INTEREST_RATE          # $1,200
loss_per_default      = LOAN_AMOUNT * (1 - RECOVERY_RATE)    # $6,000

thresholds = np.linspace(0.05, 0.95, 100)
profits, approval_rates, precision_list = [], [], []

for t in thresholds:
    approved = y_prob >= t
    tp = ((approved) & (y_test.values == 1)).sum()
    fp = ((approved) & (y_test.values == 0)).sum()
    total_approved = tp + fp

    raw_profit = tp * revenue_per_good_loan - fp * loss_per_default
    scale = N_PORTFOLIO / len(y_test)
    profits.append(raw_profit * scale)
    approval_rates.append(approved.mean() * 100)
    precision_list.append(tp / total_approved if total_approved > 0 else 0)

best_idx    = np.argmax(profits)
best_thresh = thresholds[best_idx]
best_profit = profits[best_idx]

print(f"\n── Financial Backtest Results ────────────────────────────")
print(f"  Portfolio size         : {N_PORTFOLIO:,} applicants")
print(f"  Loan amount            : ${LOAN_AMOUNT:,}")
print(f"  Annual interest rate   : {INTEREST_RATE:.0%}")
print(f"  Recovery rate          : {RECOVERY_RATE:.0%}")
print(f"  Optimal threshold      : {best_thresh:.2f}")
print(f"  Optimal approval rate  : {approval_rates[best_idx]:.1f}%")
print(f"  Expected annual profit : ${best_profit:,.0f}")
print(f"  Precision at optimum   : {precision_list[best_idx]:.1%}")

# Fig 6: Financial Backtest
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle("Financial Backtest — Simulated Credit Portfolio",
             fontsize=14, fontweight="bold", color=NAVY)

# Profit curve
axes[0].plot(thresholds, [p / 1e6 for p in profits], color=BLUE, lw=2.5)
axes[0].axvline(best_thresh, color=ORANGE, linestyle="--", lw=2,
                label=f"Optimal t = {best_thresh:.2f}")
axes[0].axhline(0, color=RED, linestyle=":", lw=1.5, alpha=0.7)
axes[0].fill_between(thresholds, [p / 1e6 for p in profits], 0,
                     where=[p > 0 for p in profits], alpha=0.12, color=GREEN)
axes[0].fill_between(thresholds, [p / 1e6 for p in profits], 0,
                     where=[p <= 0 for p in profits], alpha=0.12, color=RED)
axes[0].set(xlabel="Approval Threshold", ylabel="Expected Annual Profit ($M)",
            title="Profit Curve")
axes[0].legend()

# Approval rate curve
axes[1].plot(thresholds, approval_rates, color=GREEN, lw=2.5)
axes[1].axvline(best_thresh, color=ORANGE, linestyle="--", lw=2,
                label=f"Optimal t = {best_thresh:.2f}")
axes[1].set(xlabel="Approval Threshold", ylabel="Approval Rate (%)",
            title="Approval Rate vs. Threshold")
axes[1].legend()

# Precision vs threshold
axes[2].plot(thresholds, [p * 100 for p in precision_list], color=PURPLE, lw=2.5,
             label="Precision among approved")
axes[2].axvline(best_thresh, color=ORANGE, linestyle="--", lw=2,
                label=f"Optimal t = {best_thresh:.2f}")
axes[2].set(xlabel="Approval Threshold", ylabel="Precision (%)",
            title="Precision Among Approved Applicants")
axes[2].legend()

plt.tight_layout()
plt.savefig("fig6_backtest.png", dpi=150, bbox_inches="tight", facecolor="white")
plt.show()
print("[Saved] fig6_backtest.png")

# ── CELL 15: Summary Table ────────────────────────────────────
print("\n" + "═" * 70)
print("  FINAL MODEL PERFORMANCE SUMMARY")
print("═" * 70)

summary_data = {
    "Model": [n.replace("\n", " ") for n in cv_results.keys()],
    "CV Accuracy": [f"{cv_results[m]['accuracy'].mean():.3f} ± {cv_results[m]['accuracy'].std():.3f}"
                    for m in cv_results],
    "CV AUC-ROC":  [f"{cv_results[m]['roc_auc'].mean():.3f} ± {cv_results[m]['roc_auc'].std():.3f}"
                    for m in cv_results],
    "CV F1":       [f"{cv_results[m]['f1'].mean():.3f} ± {cv_results[m]['f1'].std():.3f}"
                    for m in cv_results],
}
summary_df = pd.DataFrame(summary_data)
print(summary_df.to_string(index=False))

print(f"""
Champion Model (XGBoost) — Hold-Out Test Set:
  AUC-ROC             : {auc:.4f}
  Gini Coefficient    : {2*auc - 1:.4f}
  KS Statistic        : {ks_stat:.4f}
  Average Precision   : {ap:.4f}
  Optimal Threshold   : {best_thresh:.2f}
  Expected Profit     : ${best_profit:,.0f}  ({N_PORTFOLIO:,} loans × ${LOAN_AMOUNT:,})

Outputs generated:
  fig0_eda.png
  fig1_model_comparison.png
  fig2_evaluation.png
  fig3a_shap_beeswarm.png
  fig3b_shap_bar.png
  fig4_waterfall_applicant_*.png  (×3)
  fig5_scorecard.png
  fig6_backtest.png
""")
```

```python
import warnings
warnings.filterwarnings("ignore")

from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from ucimlrepo import fetch_ucirepo
from xgboost import XGBClassifier
import shap

from sklearn.model_selection import train_test_split, StratifiedKFold, cross_validate
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OrdinalEncoder, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.calibration import CalibratedClassifierCV, calibration_curve
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    roc_auc_score, average_precision_score, roc_curve, precision_recall_curve,
    confusion_matrix, ConfusionMatrixDisplay, brier_score_loss, log_loss
)

OUT = Path("outputs")
FIG = OUT / "figures"
TAB = OUT / "tables"
for p in [OUT, FIG, TAB]:
    p.mkdir(parents=True, exist_ok=True)

plt.rcParams.update({
    "figure.facecolor": "white",
    "axes.facecolor": "white",
    "axes.edgecolor": "#CCCCCC",
    "axes.spines.top": False,
    "axes.spines.right": False,
    "grid.color": "#EEEEEE",
    "grid.linestyle": "--",
    "grid.alpha": 0.7,
    "font.family": "DejaVu Sans",
    "axes.titlesize": 12,
    "axes.labelsize": 10,
})

NAVY = "#1A1A2E"
BLUE = "#1F77B4"
GREEN = "#2CA02C"
ORANGE = "#FF7F0E"
RED = "#D62728"
PURPLE = "#9467BD"
GRAY = "#777777"
COLORS = [BLUE, GREEN, ORANGE, RED, PURPLE]

RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)
```

```python
# ============================================================
# 1. Load Data
# ============================================================
credit_approval = fetch_ucirepo(id=27)
X_raw = credit_approval.data.features.copy()
y_raw = credit_approval.data.targets.copy()

target_col = y_raw.columns[0]
y = (y_raw[target_col] == "+").astype(int)
y.name = "Approved"

# Working labels only. UCI intentionally anonymized the variables.
FEATURE_RENAME = {
    "A1":  "Gender_proxy",
    "A2":  "Age",
    "A3":  "Debt",
    "A4":  "Marital_Status_proxy",
    "A5":  "Bank_Customer_Type_proxy",
    "A6":  "Education_Level_proxy",
    "A7":  "A7_Protected_Class_Proxy",
    "A8":  "Years_Employed",
    "A9":  "Prior_Default",
    "A10": "Employed",
    "A11": "Credit_Score_Raw",
    "A12": "Drivers_License_proxy",
    "A13": "Citizenship_proxy",
    "A14": "Zip_Code_proxy",
    "A15": "Income",
}

VALUE_LABELS = {
    "Gender_proxy": {"b": "b", "a": "a"},
    "Prior_Default": {"t": "Yes", "f": "No"},
    "Employed": {"t": "Yes", "f": "No"},
    "Drivers_License_proxy": {"t": "Yes", "f": "No"},
}

X = X_raw.rename(columns=FEATURE_RENAME)
for col, mapping in VALUE_LABELS.items():
    if col in X.columns:
        X[col] = X[col].map(mapping).fillna(X[col])

cat_cols = X.select_dtypes(include="object").columns.tolist()
num_cols = X.select_dtypes(exclude="object").columns.tolist()

X_train_raw, X_test_raw, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=RANDOM_STATE, stratify=y
)

summary = pd.DataFrame({
    "Item": ["Observations", "Features", "Approval rate", "Total missing values", "Train observations", "Test observations"],
    "Value": [len(X), X.shape[1], f"{y.mean():.1%}", int(X.isna().sum().sum()), len(X_train_raw), len(X_test_raw)]
})
summary.to_csv(TAB / "table_data_summary.csv", index=False)
save_latex_table(summary, "table_data_summary.tex", column_format="lr")
print(summary.to_string(index=False))

print("\nNumerical features:", num_cols)
print("Categorical features:", cat_cols)
```

```python
# ============================================================
# 2. Preprocessors and Models
# ============================================================
def make_preprocessor(encoding="ordinal", drop_cols=None):
    drop_cols = set(drop_cols or [])
    use_num = [c for c in num_cols if c not in drop_cols]
    use_cat = [c for c in cat_cols if c not in drop_cols]

    num_pipe = Pipeline([
        ("imputer", SimpleImputer(strategy="median")),
        ("scaler", StandardScaler()),
    ])

    if encoding == "ordinal":
        cat_encoder = OrdinalEncoder(handle_unknown="use_encoded_value", unknown_value=-1)
    elif encoding == "onehot":
        cat_encoder = OneHotEncoder(handle_unknown="ignore", sparse_output=False)
    else:
        raise ValueError("encoding must be 'ordinal' or 'onehot'")

    cat_pipe = Pipeline([
        ("imputer", SimpleImputer(strategy="most_frequent")),
        ("encoder", cat_encoder),
    ])

    return ColumnTransformer([
        ("num", num_pipe, use_num),
        ("cat", cat_pipe, use_cat),
    ])


def xgb_model(seed=RANDOM_STATE):
    return XGBClassifier(
        n_estimators=250,
        max_depth=3,
        learning_rate=0.04,
        subsample=0.85,
        colsample_bytree=0.85,
        reg_lambda=2.0,
        reg_alpha=0.1,
        eval_metric="logloss",
        random_state=seed,
        verbosity=0,
    )

models = {
    "Logistic Regression (one-hot)": Pipeline([
        ("prep", make_preprocessor("onehot")),
        ("model", LogisticRegression(max_iter=2000, class_weight="balanced", random_state=RANDOM_STATE))
    ]),
    "Random Forest (ordinal)": Pipeline([
        ("prep", make_preprocessor("ordinal")),
        ("model", RandomForestClassifier(n_estimators=300, max_depth=5, min_samples_leaf=8, random_state=RANDOM_STATE))
    ]),
    "XGBoost (ordinal)": Pipeline([
        ("prep", make_preprocessor("ordinal")),
        ("model", xgb_model())
    ]),
    "XGBoost (one-hot)": Pipeline([
        ("prep", make_preprocessor("onehot")),
        ("model", xgb_model())
    ]),
    "SVM RBF (one-hot)": Pipeline([
        ("prep", make_preprocessor("onehot")),
        ("model", SVC(kernel="rbf", probability=True, class_weight="balanced", random_state=RANDOM_STATE))
    ]),
    "MLP (one-hot)": Pipeline([
        ("prep", make_preprocessor("onehot")),
        ("model", MLPClassifier(hidden_layer_sizes=(32, 16), max_iter=800, alpha=0.01, random_state=RANDOM_STATE))
    ]),
}

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)
scoring = ["accuracy", "roc_auc", "average_precision", "f1"]

cv_rows = []
for name, pipe in models.items():
    scores = cross_validate(pipe, X_train_raw, y_train, cv=cv, scoring=scoring, n_jobs=-1)
    cv_rows.append({
        "Model": name,
        "CV Accuracy": f"{scores['test_accuracy'].mean():.3f} ± {scores['test_accuracy'].std():.3f}",
        "CV AUC": f"{scores['test_roc_auc'].mean():.3f} ± {scores['test_roc_auc'].std():.3f}",
        "CV Avg Precision": f"{scores['test_average_precision'].mean():.3f} ± {scores['test_average_precision'].std():.3f}",
        "CV F1": f"{scores['test_f1'].mean():.3f} ± {scores['test_f1'].std():.3f}",
        "AUC Mean": scores['test_roc_auc'].mean(),
        "AUC Std": scores['test_roc_auc'].std(),
        "Accuracy Mean": scores['test_accuracy'].mean(),
        "Accuracy Std": scores['test_accuracy'].std(),
    })

cv_df = pd.DataFrame(cv_rows).sort_values("AUC Mean", ascending=False).reset_index(drop=True)
cv_df[["Model", "CV Accuracy", "CV AUC", "CV Avg Precision", "CV F1"]].to_csv(TAB / "table_cv_results.csv", index=False)
save_latex_table(
    cv_df[["Model", "CV Accuracy", "CV AUC", "CV Avg Precision", "CV F1"]],
    "table_cv_results.tex",
    column_format="lcccc"
)
print(cv_df[["Model", "CV Accuracy", "CV AUC", "CV Avg Precision", "CV F1"]].to_string(index=False))

# Figure 1: CV comparison
plot_df = cv_df.sort_values("AUC Mean", ascending=True)
fig, ax = plt.subplots(figsize=(9, 5.5))
ax.errorbar(plot_df["AUC Mean"], plot_df["Model"], xerr=plot_df["AUC Std"], fmt="o", capsize=5, lw=2)
ax.set_xlabel("5-Fold CV AUC on Training Set")
ax.set_title("Champion-Challenger Model Comparison", fontweight="bold", color=NAVY)
ax.grid(axis="x")
plt.tight_layout()
plt.savefig(FIG / "fig1_model_comparison.png", dpi=180, bbox_inches="tight")
plt.show()
```

```python
# ============================================================
# 3. Hold-Out Evaluation and Probability Calibration
# ============================================================
def ks_statistic(y_true, p):
    fpr, tpr, thr = roc_curve(y_true, p)
    idx = np.argmax(tpr - fpr)
    return float(tpr[idx] - fpr[idx]), float(thr[idx])


def evaluate_model(name, estimator, X_te=X_test_raw, y_te=y_test, threshold=0.5):
    p = estimator.predict_proba(X_te)[:, 1]
    pred = (p >= threshold).astype(int)
    auc = roc_auc_score(y_te, p)
    ks, ks_thr = ks_statistic(y_te, p)
    return {
        "Model": name,
        "AUC": auc,
        "Gini": 2 * auc - 1,
        "KS": ks,
        "KS Threshold": ks_thr,
        "Avg Precision": average_precision_score(y_te, p),
        "Brier": brier_score_loss(y_te, p),
        "Log Loss": log_loss(y_te, p),
        "Accuracy@0.5": accuracy_score(y_te, pred),
        "Precision@0.5": precision_score(y_te, pred, zero_division=0),
        "Recall@0.5": recall_score(y_te, pred, zero_division=0),
        "F1@0.5": f1_score(y_te, pred, zero_division=0),
    }

# Champion: XGBoost ordinal, chosen for strong tabular performance and SHAP compatibility.
champion_raw = models["XGBoost (ordinal)"]
champion_raw.fit(X_train_raw, y_train)

# Calibrated champion: improves probability interpretation for scorecard/backtest.
# Isotonic is flexible but can overfit small samples; sigmoid is a stable alternative.
calibrated_champion = CalibratedClassifierCV(
    estimator=models["XGBoost (ordinal)"],
    method="isotonic",
    cv=3
)
calibrated_champion.fit(X_train_raw, y_train)

# Also fit logistic baseline and one-hot XGBoost for robustness.
logit = models["Logistic Regression (one-hot)"]
logit.fit(X_train_raw, y_train)
xgb_onehot = models["XGBoost (one-hot)"]
xgb_onehot.fit(X_train_raw, y_train)

holdout_rows = [
    evaluate_model("Logistic Regression baseline", logit),
    evaluate_model("XGBoost raw", champion_raw),
    evaluate_model("XGBoost calibrated", calibrated_champion),
    evaluate_model("XGBoost one-hot robustness", xgb_onehot),
]
holdout_df = pd.DataFrame(holdout_rows)
show_cols = ["Model", "AUC", "Gini", "KS", "Avg Precision", "Brier", "Log Loss", "Accuracy@0.5", "F1@0.5"]
pretty_holdout = holdout_df[show_cols].copy()
for c in pretty_holdout.columns[1:]:
    pretty_holdout[c] = pretty_holdout[c].map(lambda x: f"{x:.3f}")
pretty_holdout.to_csv(TAB / "table_holdout_metrics.csv", index=False)
save_latex_table(pretty_holdout, "table_holdout_metrics.tex")
print(pretty_holdout.to_string(index=False))

# Use calibrated probabilities for scorecard and economic backtest.
y_prob_raw = champion_raw.predict_proba(X_test_raw)[:, 1]
y_prob = calibrated_champion.predict_proba(X_test_raw)[:, 1]

# Figure 2: ROC + calibration + confusion matrix
fig, axes = plt.subplots(1, 3, figsize=(16, 4.8))
fig.suptitle("Hold-Out Validation and Probability Calibration", fontsize=14, fontweight="bold", color=NAVY)

for name, p, color in [("Raw XGBoost", y_prob_raw, BLUE), ("Calibrated XGBoost", y_prob, GREEN), ("Logistic", logit.predict_proba(X_test_raw)[:, 1], ORANGE)]:
    fpr, tpr, _ = roc_curve(y_test, p)
    axes[0].plot(fpr, tpr, lw=2, label=f"{name}: AUC={roc_auc_score(y_test, p):.3f}")
axes[0].plot([0, 1], [0, 1], "--", color=GRAY, lw=1)
axes[0].set_title("ROC Curve")
axes[0].set_xlabel("False Positive Rate")
axes[0].set_ylabel("True Positive Rate")
axes[0].legend(fontsize=8)

for name, p, color in [("Raw XGBoost", y_prob_raw, BLUE), ("Calibrated XGBoost", y_prob, GREEN), ("Logistic", logit.predict_proba(X_test_raw)[:, 1], ORANGE)]:
    frac_pos, mean_pred = calibration_curve(y_test, p, n_bins=6, strategy="quantile")
    axes[1].plot(mean_pred, frac_pos, marker="o", lw=2, label=name)
axes[1].plot([0, 1], [0, 1], "--", color=GRAY, lw=1)
axes[1].set_title("Calibration Curve")
axes[1].set_xlabel("Mean Predicted Approval Probability")
axes[1].set_ylabel("Observed Approval Rate")
axes[1].legend(fontsize=8)

best_operating_threshold = 0.50
cm = confusion_matrix(y_test, (y_prob >= best_operating_threshold).astype(int))
ConfusionMatrixDisplay(cm, display_labels=["Denied", "Approved"]).plot(ax=axes[2], colorbar=False, cmap="Blues")
axes[2].set_title("Confusion Matrix @ 0.50")

plt.tight_layout()
plt.savefig(FIG / "fig2_holdout_calibration.png", dpi=180, bbox_inches="tight")
plt.show()
```

```python
# ============================================================
# 4. Scorecard and Reason Codes
# ============================================================
# p_approve is the calibrated probability of approval. Higher p_approve should imply a higher score.
# This corrects the common sign mistake of treating p as default probability.
PDO = 20
BASE_SCORE = 600
BASE_ODDS = 1.0
FACTOR = PDO / np.log(2)
OFFSET = BASE_SCORE - FACTOR * np.log(BASE_ODDS)

p_safe = np.clip(y_prob, 1e-6, 1 - 1e-6)
scores = np.clip(OFFSET + FACTOR * np.log(p_safe / (1 - p_safe)), 300, 850).round().astype(int)

score_bins = [300, 500, 580, 670, 740, 800, 850]
score_labels = ["Very Poor (300--499)", "Fair (500--579)", "Good (580--669)",
                "Very Good (670--739)", "Exceptional (740--799)", "Super Prime (800--850)"]
score_band = pd.cut(scores, bins=score_bins, labels=score_labels, include_lowest=True, right=False)
score_df = pd.DataFrame({
    "score": scores,
    "score_band": score_band,
    "actual_approved": y_test.values,
    "p_approve": y_prob,
})
score_stats = score_df.groupby("score_band", observed=False).agg(
    Count=("score", "count"),
    Approval_Rate=("actual_approved", "mean"),
    Avg_Score=("score", "mean"),
    Avg_P_Approve=("p_approve", "mean")
).reset_index().rename(columns={"score_band": "Score Band"})

score_stats_fmt = score_stats.copy()
score_stats_fmt["Approval_Rate"] = score_stats_fmt["Approval_Rate"].map(lambda x: "--" if pd.isna(x) else f"{x:.1%}")
score_stats_fmt["Avg_Score"] = score_stats_fmt["Avg_Score"].map(lambda x: "--" if pd.isna(x) else f"{x:.0f}")
score_stats_fmt["Avg_P_Approve"] = score_stats_fmt["Avg_P_Approve"].map(lambda x: "--" if pd.isna(x) else f"{x:.1%}")
score_stats_fmt.to_csv(TAB / "table_score_bands.csv", index=False)
save_latex_table(score_stats_fmt, "table_score_bands.tex")
print(score_stats_fmt.to_string(index=False))

fig, axes = plt.subplots(1, 3, figsize=(16, 4.8))
fig.suptitle("Credit Scorecard Outputs", fontsize=14, fontweight="bold", color=NAVY)

axes[0].hist(scores[y_test.values == 0], bins=np.linspace(300,850,35), alpha=0.65, label="Denied", density=True)
axes[0].hist(scores[y_test.values == 1], bins=np.linspace(300,850,35), alpha=0.65, label="Approved", density=True)
axes[0].set_title("Score Distribution by Outcome")
axes[0].set_xlabel("Credit Score")
axes[0].set_ylabel("Density")
axes[0].legend()

valid = score_stats.dropna(subset=["Approval_Rate"])
axes[1].bar(valid["Score Band"], valid["Approval_Rate"] * 100)
axes[1].set_title("Observed Approval Rate by Score Band")
axes[1].set_ylabel("Approval Rate (%)")
axes[1].tick_params(axis="x", rotation=35, labelsize=8)

axes[2].scatter(scores, y_prob, alpha=0.65, s=35)
axes[2].set_title("Score vs Calibrated Model Probability")
axes[2].set_xlabel("Credit Score")
axes[2].set_ylabel("Calibrated P(Approve)")

plt.tight_layout()
plt.savefig(FIG / "fig3_scorecard.png", dpi=180, bbox_inches="tight")
plt.show()
```

```python
# ============================================================
# 5. SHAP Explainability on Raw XGBoost Champion
# ============================================================
# SHAP is computed on the raw XGBoost model for transparent tree-level attribution.
# Calibration is used for final probabilities, not for feature-attribution mechanics.
prep = champion_raw.named_steps["prep"]
xgb = champion_raw.named_steps["model"]
X_train_proc = prep.transform(X_train_raw)
X_test_proc = prep.transform(X_test_raw)
feature_names = num_cols + cat_cols

explainer = shap.TreeExplainer(xgb)
shap_values = explainer(X_test_proc)
shap_values.feature_names = feature_names

plt.figure(figsize=(9.5, 6))
shap.plots.beeswarm(shap_values, max_display=14, show=False)
plt.title("SHAP Feature Importance — Raw XGBoost Champion", fontweight="bold", color=NAVY)
plt.tight_layout()
plt.savefig(FIG / "fig4_shap_beeswarm.png", dpi=180, bbox_inches="tight")
plt.show()

# Select examples programmatically so figures never point to missing files.
# clear approval: high p; clear denial: low p; borderline: closest to chosen threshold 0.5.
example_ids = {
    "clear_approval": int(np.argmax(y_prob)),
    "clear_denial": int(np.argmin(y_prob)),
    "borderline": int(np.argmin(np.abs(y_prob - 0.5))),
}

reason_code_rows = []
for label, idx in example_ids.items():
    plt.figure(figsize=(10, 5.5))
    shap.plots.waterfall(shap_values[idx], max_display=12, show=False)
    decision = "APPROVED" if y_prob[idx] >= 0.5 else "DECLINED"
    plt.title(f"{label.replace('_',' ').title()} | P(Approve)={y_prob[idx]:.1%} | Score={scores[idx]} | {decision}",
              fontweight="bold", color=NAVY)
    plt.tight_layout()
    fname = f"fig5_waterfall_{label}.png"
    plt.savefig(FIG / fname, dpi=180, bbox_inches="tight")
    plt.show()

    # Most negative SHAP values reduce approval probability. Use original variable labels as reason-code candidates.
    sv = pd.DataFrame({"Feature": feature_names, "SHAP": shap_values[idx].values})
    adverse = sv.sort_values("SHAP").head(3)
    for rank, row in enumerate(adverse.itertuples(index=False), 1):
        reason_code_rows.append({
            "Applicant Type": label.replace("_", " ").title(),
            "P(Approve)": f"{y_prob[idx]:.1%}",
            "Score": int(scores[idx]),
            "Decision": decision,
            "Reason Rank": rank,
            "Feature": row.Feature,
            "SHAP Value": f"{row.SHAP:.3f}",
        })

reason_codes = pd.DataFrame(reason_code_rows)
reason_codes.to_csv(TAB / "table_reason_codes.csv", index=False)
save_latex_table(reason_codes, "table_reason_codes.tex")
print(reason_codes.to_string(index=False))
```

```python
# ============================================================
# 6. Financial Backtests: Flat and Risk-Based Economics
# ============================================================
FUNDING_COST = 0.04
OPERATING_COST = 150
N_PORTFOLIO = 10_000

flat_assumptions = {
    "loan_amount": 10_000,
    "apr": 0.12,
    "recovery": 0.40,
}

# Risk-based policy by calibrated score. These are modeling assumptions, not observed UCI data.
def assign_risk_terms(score):
    if score >= 700:
        return pd.Series({"Risk Band": "Prime", "Loan Amount": 12_000, "APR": 0.09, "Recovery": 0.55})
    elif score >= 620:
        return pd.Series({"Risk Band": "Near-prime", "Loan Amount": 10_000, "APR": 0.13, "Recovery": 0.45})
    elif score >= 560:
        return pd.Series({"Risk Band": "Subprime", "Loan Amount": 7_000, "APR": 0.20, "Recovery": 0.35})
    else:
        return pd.Series({"Risk Band": "Deep subprime", "Loan Amount": 4_000, "APR": 0.27, "Recovery": 0.25})

terms = pd.DataFrame([assign_risk_terms(s) for s in scores])

def loan_profit(y_actual, loan_amount, apr, recovery, scenario_multiplier=1.0):
    # y_actual=1 means good/approved outcome; y_actual=0 means bad/denied outcome in the historical label.
    good_profit = loan_amount * (apr - FUNDING_COST) - OPERATING_COST
    default_loss = loan_amount * (1 - recovery) * scenario_multiplier + OPERATING_COST
    return np.where(y_actual == 1, good_profit, -default_loss)

def run_backtest(prob, y_actual, thresholds, mode="flat", scenario_multiplier=1.0):
    rows = []
    for t in thresholds:
        approved = prob >= t
        if mode == "flat":
            per_loan_profit = loan_profit(
                y_actual,
                flat_assumptions["loan_amount"],
                flat_assumptions["apr"],
                flat_assumptions["recovery"],
                scenario_multiplier=scenario_multiplier,
            )
        elif mode == "risk_based":
            per_loan_profit = loan_profit(
                y_actual,
                terms["Loan Amount"].values,
                terms["APR"].values,
                terms["Recovery"].values,
                scenario_multiplier=scenario_multiplier,
            )
        else:
            raise ValueError

        raw_profit = per_loan_profit[approved].sum()
        scale = N_PORTFOLIO / len(y_actual)
        approved_n = approved.sum()
        approved_good = ((approved) & (y_actual == 1)).sum()
        approved_bad = ((approved) & (y_actual == 0)).sum()
        rows.append({
            "Threshold": t,
            "Approval Rate": approved.mean(),
            "Bad Rate Among Approved": approved_bad / approved_n if approved_n else np.nan,
            "Precision Among Approved": approved_good / approved_n if approved_n else np.nan,
            "Expected Profit": raw_profit * scale,
        })
    return pd.DataFrame(rows)

thresholds = np.linspace(0.05, 0.95, 181)
flat_bt = run_backtest(y_prob, y_test.values, thresholds, mode="flat")
risk_bt = run_backtest(y_prob, y_test.values, thresholds, mode="risk_based")

# Scenario stress test on risk-based economics.
scenario_defs = {
    "Base": 1.00,
    "Mild Stress": 1.25,
    "Severe Stress": 1.50,
}
scenario_rows = []
for scen, mult in scenario_defs.items():
    bt = run_backtest(y_prob, y_test.values, thresholds, mode="risk_based", scenario_multiplier=mult)
    best = bt.loc[bt["Expected Profit"].idxmax()]
    scenario_rows.append({
        "Scenario": scen,
        "Default Loss Multiplier": f"{mult:.2f}x",
        "Optimal Threshold": f"{best['Threshold']:.2f}",
        "Approval Rate": f"{best['Approval Rate']:.1%}",
        "Bad Rate Among Approved": f"{best['Bad Rate Among Approved']:.1%}",
        "Expected Profit": f"${best['Expected Profit']:,.0f}",
    })
scenario_df = pd.DataFrame(scenario_rows)
scenario_df.to_csv(TAB / "table_backtest_scenarios.csv", index=False)
save_latex_table(scenario_df, "table_backtest_scenarios.tex")
print(scenario_df.to_string(index=False))

# Terms by score band table
terms_table = pd.concat([score_df[["score"]].reset_index(drop=True), terms.reset_index(drop=True)], axis=1)
policy_summary = terms_table.groupby("Risk Band", observed=False).agg(
    Count=("score", "count"),
    Min_Score=("score", "min"),
    Max_Score=("score", "max"),
    Avg_Loan=("Loan Amount", "mean"),
    Avg_APR=("APR", "mean"),
    Avg_Recovery=("Recovery", "mean"),
).reset_index()
policy_fmt = policy_summary.copy()
policy_fmt["Avg_Loan"] = policy_fmt["Avg_Loan"].map(lambda x: f"${x:,.0f}")
policy_fmt["Avg_APR"] = policy_fmt["Avg_APR"].map(lambda x: f"{x:.1%}")
policy_fmt["Avg_Recovery"] = policy_fmt["Avg_Recovery"].map(lambda x: f"{x:.1%}")
policy_fmt.to_csv(TAB / "table_risk_based_terms.csv", index=False)
save_latex_table(policy_fmt, "table_risk_based_terms.tex")

# Figure: profit curves and risk metrics
fig, axes = plt.subplots(1, 3, figsize=(16, 4.8))
fig.suptitle("Financial Backtest: Flat vs Risk-Based Lending Economics", fontsize=14, fontweight="bold", color=NAVY)
for name, bt, color in [("Flat $10k Loan", flat_bt, BLUE), ("Risk-Based Pricing/Limit", risk_bt, GREEN)]:
    axes[0].plot(bt["Threshold"], bt["Expected Profit"] / 1e6, lw=2.2, label=name)
    best = bt.loc[bt["Expected Profit"].idxmax()]
    axes[0].scatter(best["Threshold"], best["Expected Profit"] / 1e6, s=70)
axes[0].axhline(0, color=GRAY, linestyle="--", lw=1)
axes[0].set_title("Expected Profit")
axes[0].set_xlabel("Approval Threshold")
axes[0].set_ylabel("Profit ($M, scaled to 10k applicants)")
axes[0].legend(fontsize=8)

axes[1].plot(risk_bt["Threshold"], risk_bt["Approval Rate"] * 100, color=GREEN, lw=2.2)
axes[1].set_title("Approval Rate")
axes[1].set_xlabel("Approval Threshold")
axes[1].set_ylabel("Approval Rate (%)")

axes[2].plot(risk_bt["Threshold"], risk_bt["Bad Rate Among Approved"] * 100, color=RED, lw=2.2)
axes[2].set_title("Bad Rate Among Approved")
axes[2].set_xlabel("Approval Threshold")
axes[2].set_ylabel("Bad Rate (%)")

plt.tight_layout()
plt.savefig(FIG / "fig6_backtest_risk_based.png", dpi=180, bbox_inches="tight")
plt.show()

flat_best = flat_bt.loc[flat_bt["Expected Profit"].idxmax()]
risk_best = risk_bt.loc[risk_bt["Expected Profit"].idxmax()]
backtest_summary = pd.DataFrame([
    {"Policy": "Flat $10k loan", "Optimal Threshold": f"{flat_best['Threshold']:.2f}", "Approval Rate": f"{flat_best['Approval Rate']:.1%}", "Bad Rate Among Approved": f"{flat_best['Bad Rate Among Approved']:.1%}", "Expected Profit": f"${flat_best['Expected Profit']:,.0f}"},
    {"Policy": "Risk-based pricing/limit", "Optimal Threshold": f"{risk_best['Threshold']:.2f}", "Approval Rate": f"{risk_best['Approval Rate']:.1%}", "Bad Rate Among Approved": f"{risk_best['Bad Rate Among Approved']:.1%}", "Expected Profit": f"${risk_best['Expected Profit']:,.0f}"},
])
backtest_summary.to_csv(TAB / "table_backtest_summary.csv", index=False)
save_latex_table(backtest_summary, "table_backtest_summary.tex")
print(backtest_summary.to_string(index=False))
```

```python
# ============================================================
# 7. Fairness / A7 Proxy Ablation
# ============================================================
def max_group_approval_gap(prob, threshold=0.5, group_col="A7_Protected_Class_Proxy", min_n=5):
    temp = pd.DataFrame({
        "group": X_test_raw[group_col].astype(str).fillna("Missing"),
        "approved_model": prob >= threshold,
    })
    rates = temp.groupby("group").agg(n=("approved_model", "size"), approval_rate=("approved_model", "mean"))
    rates = rates[rates["n"] >= min_n]
    if len(rates) < 2:
        return np.nan, rates
    return float(rates["approval_rate"].max() - rates["approval_rate"].min()), rates

xgb_no_a7 = Pipeline([
    ("prep", make_preprocessor("ordinal", drop_cols=["A7_Protected_Class_Proxy"])),
    ("model", xgb_model())
])
xgb_no_a7.fit(X_train_raw.drop(columns=[], errors="ignore"), y_train)

p_with_a7 = y_prob
p_without_a7 = xgb_no_a7.predict_proba(X_test_raw)[:, 1]

with_gap, with_rates = max_group_approval_gap(p_with_a7)
without_gap, without_rates = max_group_approval_gap(p_without_a7)

fairness_rows = []
for name, p, gap in [("Calibrated XGBoost with A7", p_with_a7, with_gap), ("XGBoost excluding A7", p_without_a7, without_gap)]:
    auc = roc_auc_score(y_test, p)
    ks, _ = ks_statistic(y_test, p)
    bt = run_backtest(p, y_test.values, thresholds, mode="risk_based")
    best = bt.loc[bt["Expected Profit"].idxmax()]
    fairness_rows.append({
        "Model": name,
        "AUC": f"{auc:.3f}",
        "KS": f"{ks:.3f}",
        "Optimal Threshold": f"{best['Threshold']:.2f}",
        "Expected Profit": f"${best['Expected Profit']:,.0f}",
        "Max A7 Approval Gap": "--" if pd.isna(gap) else f"{gap:.1%}",
    })
fairness_df = pd.DataFrame(fairness_rows)
fairness_df.to_csv(TAB / "table_fairness_ablation.csv", index=False)
save_latex_table(fairness_df, "table_fairness_ablation.tex")
print(fairness_df.to_string(index=False))

# Group approval-rate figure
rates_plot = pd.concat([
    with_rates.assign(Model="With A7"),
    without_rates.assign(Model="Without A7")
]).reset_index()
fig, ax = plt.subplots(figsize=(9, 4.8))
for label, data in rates_plot.groupby("Model"):
    ax.plot(data["group"], data["approval_rate"] * 100, marker="o", lw=2, label=label)
ax.set_title("Model Approval Rate by A7 Proxy Group", fontweight="bold", color=NAVY)
ax.set_ylabel("Approval Rate (%)")
ax.set_xlabel("A7 Proxy Group (anonymized)")
ax.tick_params(axis="x", rotation=30)
ax.legend()
plt.tight_layout()
plt.savefig(FIG / "fig7_fairness_ablation.png", dpi=180, bbox_inches="tight")
plt.show()
```
